#### 函数：  
tm1.dimensions.get_all_names()->返回所有维度的名称列表  
tm1.dimensions.get(dimension_name='')->返回某一维度的实例
#### 循环：  
for hier in tm1.dimensions.get(dimension_name=''):->第一次循环外层为hier层级对象  
    for ele in hier:->第二层循环hier中的元素对象  
    for sub in hier.subsets:->第二层循环hier的subsets属性中的子集名称  
    for parent, child in hier.edges:->第二层循环hier的edges属性中的父子层级元素  
#### 属性：  
hier.name->hier名称 
hier.subsets->hier子集  
hier.edges->hier的edges字典  
ele.name->元素名称  
ele.index->元素索引 
ele.element_type->元素类型
    


```python
from TM1py.Services import TM1Service

# Connection to TM1. Needs Address, Port, Credentials, and SSL
with TM1Service(address="localhost", port=30011, ssl=True, user="admin", password="apple") as tm1:
    # get random dimension from the model
    dimension_names = tm1.dimensions.get_all_names()
   
    #print(dimension_names)
    dimension = tm1.dimensions.get(dimension_name=dimension_names[5])
    
    
    # iterate through hierarchies
    for hierarchy in dimension:
        print('Hierarchy Name: {}'.format(hierarchy.name))
        # iterate through Elements in hierarchy
        for element in hierarchy:
            print('Element Name: {} Index: {} Type: {} Attr: {}'.format(element.name, str(element.index), element.element_type,element.element_attributes))
            
        # iterate through Subsets
        for subset in hierarchy.subsets:
            print('Subset Name: {}'.format(subset))
        # iterate through Edges
        for parent, child in hierarchy.edges:
            print("Parent Name: {}, Component Name: {}".format(parent, child))

        # print the default member
        print('Default Member: {}'.format(hierarchy.default_member))
```
