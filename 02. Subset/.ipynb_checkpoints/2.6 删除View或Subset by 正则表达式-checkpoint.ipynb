#### 函数：  
返回instance中所有cube名称 
tm1.cubes.get_all_names()  
tm1.cubes.get_all_names()[0]->返回第一个cube实例  
返回某个cube实例的视图实例  
View1,View2=tm1.cubes.views.get_all(cube_name='')->返回的是两个实例，第一个是私有视图，第二个是公共视图  
tm1.cubes.views.get_all(cube_name='')[0]->查看私有视图实例显示成列表的信息    
tm1.cubes.views.get_all(cube_name='')[1]->查看公有视图实例显示成列表的信息  
视图实例的属性  
View2.name->返回视图的名字  
删除视图  
tm1.cubes.views.delete(cube_name='', view_name='', private=True)  
返回instance中所有维度名称  
tm1.dimensions.get_all_names()  
获取某个维度中所有的subset名称  
tm1.dimensions.subsets.get_all_names(dimension_name='', hierarchy_name='')  
删除某个维度中的某个subset  
tm1.dimensions.subsets.delete(dimension_name='', subset_name='')  
##### 正则函数  
re.match('正则表达式','文本',re.IGNORECASE)->如果适配则返回True否则False,IGNORECASE忽略大小写  

```python
import re

from TM1py.Services import TM1Service

with TM1Service(address="localhost", port=30011, ssl=True, user="admin", password="apple") as tm1:
    # 正则表达式 以'temp_'或'test_'开头的字符串，^代表开始,*代表结束,.代表任意字符
    regex_list = ['^}temp_.*', '^test_.*']

    # 遍历所有Cube
    cubes = tm1.cubes.get_all_names()
    for cube in cubes:
        #返回某cube的公众与私人视图实例列表
        private_views, public_views = tm1.cubes.views.get_all(cube_name=cube)
        
        for view in private_views:
            for regex in regex_list:
                # 正则适配
                if re.match(regex, view.name, re.IGNORECASE):
                    tm1.cubes.views.delete(cube_name=cube, view_name=view.name, private=True)
        for view in public_views:
            for regex in regex_list:
                if re.match(regex, view.name, re.IGNORECASE):
                    tm1.cubes.views.delete(cube_name=cube, view_name=view.name, private=False)

    # iterate through dimensions
    dimensions = tm1.dimensions.get_all_names()
    for dimension in dimensions:
        subsets = tm1.dimensions.subsets.get_all_names(dimension_name=dimension, hierarchy_name=dimension)
        for subset in subsets:
            for regex in regex_list:
                if re.match(regex, subset, re.IGNORECASE):
                    tm1.dimensions.subsets.delete(dimension_name=dimension, subset_name=subset)
```
