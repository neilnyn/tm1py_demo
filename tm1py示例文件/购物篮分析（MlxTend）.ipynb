### 读取数据

```python
import pandas as pd
```

```python
zl_order_data=pd.read_csv(r"C:/TM1py/mba.csv")
```

```python
zl_order_data['Region'].unique()
```

```python
pYear_start=2020
pMonth_start=1
pYear_end=2021
pMonth_end=12
pOU="2"
pBU="hangzhou"
pChannel="Amazon"
pRegion="uk"
```

```python
zl_order_data=zl_order_data[(zl_order_data["yyyymm"].str.slice(0,4).astype("int")>=pYear_start)&(zl_order_data["yyyymm"].str.slice(5,7).astype("int")>=pMonth_start)
                            &(zl_order_data["yyyymm"].str.slice(0,4).astype("int")<=pYear_end)&(zl_order_data["yyyymm"].str.slice(5,7).astype("int")<=pMonth_end)
                            &(zl_order_data["OU"].astype("str")==pOU)
                            &(zl_order_data["BU"]==pBU)
                            &(zl_order_data["Channel"]==pChannel)
                            &(zl_order_data["Region"]==pRegion)].drop(["yyyymm","OU","BU","Channel","Region"],axis=1)
```

```python
zl_order_data
```

```python
zl_order_data.reset_index(drop=True,inplace=True)
```

```python
zl_order_multi_order=zl_order_data[zl_order_data.groupby("order_id")["order_id"].transform("count")>1]
```

```python
zl_order_multi_order_drop=zl_order_multi_order.drop_duplicates()
```

```python
zl_order_multi_order_drop.reset_index(drop=True,inplace=True)
```

```python
zl_order_multi_order_drop
```

#### 转换为交叉表

```python
modeldata=pd.crosstab(index=zl_order_multi_order_drop["order_id"],columns=zl_order_multi_order_drop["lsku"])
#新版本要转换1->True 0->False
def func(col):
    result=col.copy()
    result_bool=pd.Series(result==1)
    col=result_bool
    return col
```

```python
modeldata_bool=modeldata.apply(func,axis=1)
```

```python
modeldata_bool
```

#### 生成频繁项集

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
fre_set=apriori(modeldata_bool,min_support=0.0001,use_colnames=True,low_memory=True)
```

#### 关联规则

```python
rules=association_rules(fre_set,metric="confidence",min_threshold=0.8)
```

```python
round(rules,3)
```

### 写入TM1

```python
rules_tm1=rules.copy()
```

```python
rules_tm1.reset_index(inplace=True)
rules_tm1.rename(columns={"index":"Line Item"},inplace=True)
rules_tm1["Line Item"]=rules_tm1["Line Item"].apply(lambda x:"{:0>4d}".format(x)).astype('str')
```

```python
rules_tm1_melt=pd.melt(rules_tm1,id_vars=["Line Item"],var_name=["M ZL Sales Market Basket Analysis"],value_name="Value")
```

```python
rules_tm1_melt["Scenario"]="ACT"
rules_tm1_melt["SA Year"]="Unspecified Year"
rules_tm1_melt["SA Month"]="Unspecified Month"
rules_tm1_melt["Version"]="WIP"
rules_tm1_melt["COA Company"]="{:0>3d}".format(int(pOU))
rules_tm1_melt["Business Unit"]=pBU
rules_tm1_melt["Region"]=pRegion
rules_tm1_melt["Channel"]=pChannel
rules_tm1_melt["Data Source ZL Sales Market Basket Analysis"]=str(pYear_start)+str("{:0>2d}".format(pMonth_start))+"_"+str(pYear_end)+str("{:0>2d}".format(pMonth_end))
```

```python
columns=["Scenario","SA Year","SA Month","Version","COA Company","Business Unit","Region","Channel","Line Item","Data Source ZL Sales Market Basket Analysis","M ZL Sales Market Basket Analysis","Value"]
rules_tm1_melt_ready=rules_tm1_melt.reindex(columns=columns)
```
