```python
cnt=0
for c in df_read_c:
    if cnt<1:
        print(c.groupby(by='website').sum())
    else:
        break
    cnt+=1    
```

```python
cnt=0
for c in df_read_c:
    if cnt==0:
        df_all=pd.DataFrame({'website':[],'quantity':[]})
        df_all.reset_index(inplace=True)
        df_all.drop('index',axis=1,inplace=True)
    df_c=c.groupby(by='website').sum()
    df_c.reset_index(inplace=True)
    df_all=pd.concat([df_all,df_c])
    cnt+=1
```

```python
file=df_all.groupby(by='website').sum().sort_values(by='quantity',ascending=False)
```
