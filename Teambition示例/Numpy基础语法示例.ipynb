```python
executive=pd.read_csv(r'C:/数据练习/5数据的科学运算/executive.csv',chunksize=100)
#executive.get_chunk(102)
```

```python
#分块，累加
number=0
for c in executive:
    number+=c['age'].sum()
    
```

```python
for c in executive:
    df_c=c.groupby(by='sex').mean()
    df_c.reset_index(inplace=True)
    print(df_c)
```

```python
cnt=0
for c in executive:
    if cnt==0:
        df_all=pd.DataFrame({'sex':[],'age':[]})
        df_all.reset_index(inplace=True)
        df_all.drop('index',axis=1,inplace=True)
    df_c=c.groupby(by='sex').mean()
    df_c.reset_index(inplace=True)
    df_all=pd.concat([df_all,df_c])
    cnt+=1
```
