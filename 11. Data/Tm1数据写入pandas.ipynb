```python
from TM1py.Services import TM1Service
from TM1py.Utils import Utils

with TM1Service(address="localhost", port=30011, ssl=True, user="admin", password="apple") as tm1:
    # define MDX Query
    mdx = '''SELECT 
          NON EMPTY 
           {[Month].[M01],[Month].[M02],[Month].[M03],[Month].[M04],[Month].[M05],[Month].[M06]} 
            PROPERTIES [Month].[Month].[Long Name]  ON COLUMNS , 
            NON EMPTY 
           {TM1DRILLDOWNMEMBER({[Account].[Account].[Net Income],[Account].[Account].[Operating Profit],[Account].[Account].[Gross Margin],[Account].[Account].[Operating Expenses],[Account].[Account].[Other Income and Expense],[Account].[Account].[Taxes]}, ALL, RECURSIVE )} 
            ON ROWS 
            FROM [General Ledger] 
            WHERE 
              (
               [Year].[Year].[FY11],
               [Version].[Version].[Budget],
               [Organisation].[Organisation].[Canadian Division],
               [Department].[Department].[Research and Development],
               [GL Measure].[GL Measure].[Amount]
              )'''

    # Get data from P&L cube through MDX
    pnl_data = tm1.cubes.cells.execute_mdx(mdx)

    # Build pandas DataFrame fram raw cellset data
    df = Utils.build_pandas_dataframe_from_cellset(pnl_data)

    print(df)

    # Calculate Statistical measures for dataframe
    print(df.describe())
```
