#### 函数：  
Process(name='pro_name',datasource_type='ASCII/TM1CubeView/ODBC')->创建process实例  
tm1.processes.create(Process(name='',datasource_type=''))->更新process实例  
#### 方法：  
Process(name='',datasource_type='').add_variable('Variable', 'String/Numeric')->创建变量  
#### 属性：  
Process(name='',datasource_type='').prolog_procedure='code'->将TI文本导入prolog  
Process(name='',datasource_type='').metadata_procedure='code'->将TI文本导入metadata  
.  
.  
.  

```python
from TM1py.Objects import Process
from TM1py.Services import TM1Service
```

```python
"""
Create a TI process in TM1
"""


# connection to TM1 Server
with TM1Service(address="localhost", port=30011, ssl=True, user="admin", password="apple") as tm1:
    process_name = 'TM1py process'
    #cCubeSource='Employee'
    # create new Process in python
    p_ascii = Process(name=process_name,
                      datasource_type='TM1CubeView')
                      #datasource_data_source_name_for_server=cCubeSource,
                      #datasource_data_source_name_for_client=cCubeSource)
    
    
    
    
    
    
    # variables
    p_ascii.add_variable('vVersion', 'String')
    p_ascii.add_variable('vYear', 'String')
    p_ascii.add_variable('vMonth', 'String')
    p_ascii.add_variable('vOrg', 'String')
    p_ascii.add_variable('vDep', 'String')
    p_ascii.add_variable('vEmployee', 'String')
    p_ascii.add_variable('vMeasure', 'String')
    p_ascii.add_variable('vValue', 'Numeric')
    # parameters
    #p_ascii.add_parameter(name='pCompanyCode', prompt='', value='DE04')
    # code
    p_ascii.prolog_procedure = '''
    #临时view前缀
    cViewSrcPrefix= '}Temp_';
    cProcessName=getprocessname();
    #临时view名称
    cViewSrc = cViewSrcPrefix | cProcessName;
    #临时subset名称
    cSubSrc = cViewSrc;
    #Cube名字
    cCubeSource='Employee';

    #建空View
    If (ViewExists(cCubeSource, cViewSrc) = 0);
        ViewCreate(cCubeSource, cViewSrc);
    Endif;



    #某维度1
    sDim = 'Version';
    If (SubsetExists(sDim, cSubSrc) = 0);
        SubsetCreate(sDim, cSubSrc);
    Endif;
    SubsetDeleteAllElements(sDim, cSubSrc);
    SubsetElementInsert(sDim, cSubSrc, 'Actual', 1);
    ViewSubsetAssign(cCubeSource, cViewSrc, sDim, cSubSrc);
    ViewRowDimensionSet(cCubeSource, cViewSrc, sDim, 1);


    #某维度2
    sDim = 'Year';
    If (SubsetExists(sDim, cSubSrc) = 0);
        SubsetCreate(sDim, cSubSrc);
    Endif;
    SubsetDeleteAllElements(sDim, cSubSrc);
    SubsetElementInsert(sDim, cSubSrc, 'FY13', 1);
    ViewSubsetAssign(cCubeSource, cViewSrc, sDim, cSubSrc);
    ViewRowDimensionSet(cCubeSource, cViewSrc, sDim, 2);

 
    #某维度3
    sDim = 'Month';

    ViewSubsetAssign(cCubeSource, cViewSrc, sDim, 'Level 0');
    ViewRowDimensionSet(cCubeSource, cViewSrc, sDim, 3);

 
    #某维度4
    sDim = 'Organisation';
    MDX='{Tm1filterbylevel({Tm1subsetall([Organisation])},0)}';
    If (SubsetExists(sDim, cSubSrc) = 0);
        SubsetCreatebyMDX(cSubSrc, MDX);
    Endif;
    ViewSubsetAssign(cCubeSource, cViewSrc, sDim, cSubSrc);
    ViewRowDimensionSet(cCubeSource, cViewSrc, sDim, 4);


    #某维度5
    sDim = 'Department';
    MDX='{Tm1filterbylevel({Tm1subsetall([Department])},0)}';
    If (SubsetExists(sDim, cSubSrc) = 0);
        SubsetCreatebyMDX(cSubSrc, MDX);
    Endif;
    ViewSubsetAssign(cCubeSource, cViewSrc, sDim, cSubSrc);
    ViewRowDimensionSet(cCubeSource, cViewSrc, sDim, 5);

    #某维度6
    sDim = 'Employee';
    MDX='{Tm1filterbylevel({Tm1subsetall([Employee])},0)}';
    If (SubsetExists(sDim, cSubSrc) = 0);
        SubsetCreatebyMDX(cSubSrc, MDX);
    Endif;
    ViewSubsetAssign(cCubeSource, cViewSrc, sDim, cSubSrc);
    ViewRowDimensionSet(cCubeSource, cViewSrc, sDim, 6);

    #某维度7
    sDim = 'Employee Measure';
    If (SubsetExists(sDim, cSubSrc) = 0);
        SubsetCreate(sDim, cSubSrc);
    Endif;
    SubsetDeleteAllElements(sDim, cSubSrc);
    SubsetElementInsert(sDim, cSubSrc, 'Base Salary', 1);
    ViewSubsetAssign(cCubeSource, cViewSrc, sDim, cSubSrc);
    ViewRowDimensionSet(cCubeSource, cViewSrc, sDim, 7);


    #忽略c Level
    ViewExtractSkipCalcsSet(cCubeSource, cViewSrc, 0);
    #忽略rule
    ViewExtractSkipRuleValuesSet(cCubeSource, cViewSrc, 1);
    #抑制0
    ViewExtractSkipZeroesSet(cCubeSource, cViewSrc, 1);
    #固定套路四句话
    DataSourceType = 'VIEW';
    DatasourceNameForServer = cCubeSource;
    DatasourceNameForClient = cCubeSource;
    DatasourceCubeView = cViewSrc;'''
    p_ascii.data_procedure='''
    sMeasure=numbertostring(vValue);
    ASCIIOutput('C:\DBs_EDU\cw_af\Employee.csv', vVersion, vYear, vMonth,vOrg,vDep,vEmployee,vMeasure,sMeasure);'''
    # create process on TM1 Server
    tm1.processes.create(p_ascii)
```
