#### 函数：  
tm1.processes.get('TI_name')->获取某TI的实例  
#### 属性：  
tm1.processes.get('').parameters->获取参数  
tm1.processes.get('').variables->获取变量  
tm1.processes.get('').prolog_procedure->获取prolog code   
tm1.processes.get('').metadata_procedure->获取meta code  
tm1.processes.get('').data_procedure->获取data code  
tm1.processes.get('').epilog_procedure->获取epilog code  

```python
"""
Query a Process from the TM1 model
"""

from TM1py.Services import TM1Service


# connection to TM1 Server
with TM1Service(address="localhost", port=30011, ssl=True, user="admin", password="apple") as tm1:
    # read Process
    p = tm1.processes.get('Cub.Employee.Export.CSV')
    print(p.datasource_type)
    # print variables, parameters, ...
    print('Parameters: \r\n' + str(p.parameters))
    print('Variables: \r\n' + str(p.variables))
    print('Prolog: \r\n' + str(p.prolog_procedure))
    print('Metadata: \r\n' + str(p.metadata_procedure))
    print('Data: \r\n' + str(p.data_procedure))
    print('Epilog: \r\n' + str(p.epilog_procedure))
```
